```python
from pandas import DataFrame
import pandas as pd
import numpy as np
```

```python
df = DataFrame({'key1': ['a', 'a', 'b', 'b', 'a'],
                'key2': ['one', 'two', 'one', 'two', 'one'],
                'data1': np.random.randn(5),
                'data2': np.random.randn(5)})
df
```

```python
def peak_to_peak(arr):
    if arr.dtype == np.float64:
        return arr.max() - arr.min()

grouped = df.groupby('key1')
grouped.head()
grouped[['data1', 'data2', 'key1']].agg(peak_to_peak)
```

```python
grouped.describe()
```

```python
# add more than one aggregation
grouped.agg([np.mean, np.sum])
```

```python
tips = pd.read_csv('tips.csv')
del tips['Unnamed: 0']
tips.head()
```

```python
tips['tip_pct'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
grouped = tips.groupby(['sex', 'smoker'])
grouped_pct = grouped['tip_pct']
grouped_pct.agg('mean')
```

```python
# using multiple aggregations
grouped_pct.agg(['mean', 'std', peak_to_peak])
```

```python
# if one wants to name resulting columns differently
grouped.agg([('my_mean', 'mean'), ('my_std', np.std)])
```

```python
# using multiple functions
functions = ['count', 'mean', 'max']
result = grouped['tip_pct', 'total_bill'].agg(functions)
result
# but if one wants more exotic names
ftuples = [('Durchschnitt', 'mean'), ('Abweichung', np.var)]
grouped['tip_pct', 'total_bill'].agg(ftuples)
```

```python
# applying different functions to different columns
grouped.agg({'tip' : [('tip_max', np.max), ('tip_sum', sum)], 'size' : 'sum'})
```

```python
# disabling hierachical index creation
tips.groupby(['sex', 'smoker'], as_index=False).mean()
```

```python
# Apply: general split-apply-combine
# let's select top 5 tips from each group
def top(df, n=5, column='tip_pct'):
    return df.sort_index(by=column)[-n:]

top(tips, n=4)
```

```python
# let's call groupby for smoker and apply top function
# and under the hood: the top function was called on each piece
# of the DataFraem, then the results are glued together using pandas.concat,
# labeling the pieces with the group names. The result has a hierachical index
# whose inner level contains index values from the original DataFrame
tips.groupby('smoker').apply(top)
```

```python
# pass arguments to the apply function
tips.groupby(['smoker', 'day']).apply(top, n=1, column='total_bill')
```

```python
# supressing the group key
tips.groupby('smoker', group_keys=False).apply(top)
```

```python
frame = DataFrame({'data1': np.random.randn(1000), 
                   'data2': np.random.randn(1000)})
factor = pd.cut(frame.data1, 4)
factor[:10]
```

```python
def get_stats(group):
    return {'min': group.min(), 'max': group.max(),
            'count': group.count(), 'mean': group.mean()}

grouped = frame.data2.groupby(factor)
grouped.apply(get_stats).unstack()
```

```python
# to compute equal-size buckets based on sample
# quantile one must use qcut
grouping = pd.qcut(frame.data1, 10, labels=False)
grouped = frame.data2.groupby(grouping)
grouped.apply(get_stats).unstack()
```
