```python
import pandas as pd
```

```python
from datetime import date, timedelta

def excel_num2date(excel_num):
    if type(excel_num) is datetime:
        return excel_num
    
    return pd.to_datetime(
        date(1900, 1, 1) + timedelta(int(excel_num) - 2))
```

```python
path = '/home/maciej/projects/temp/SOV_data_Final.xlsx'

# add prop (proportion) column
def add_sov(group):
    # Integer division floors
    count = group.name_count.astype(float)
    
    group['SOV'] = count / count.sum()
    return group

def make_stuff_for_month(month):
    # -- radian6 data
    radian6 = pd.read_excel(path, month, parse_cols='A:C', names=['date', 'name', 'name_count'])
    radian6 = radian6[radian6.date.notnull()]  # filter out empty rows 
    #words.date = words.date.apply(excel_num2date)  # make sure that date is "seen" correctly
    radian6 = radian6.groupby('date').apply(add_sov)
    radian6.name = radian6.name.apply(lambda x: x[0].upper() + x[1:])
    
    assert np.allclose(radian6.groupby('date').SOV.sum(), 1)  # sanity check
    
    # -- stats data
    snames = ["date", "viral reach", "organic reach", "paid reach", 
              "daily engaged", "day", "lifetime likes", "daily new like",
              "daily new unlike", "weekly people talking about this", 
              "weekly page engagement rate"]
    
    stats = pd.read_excel(path, month, parse_cols='F:P', names=snames)
    stats = stats[stats.date.notnull()]
    #stats.date = stats.date.apply(pd.to_datetime)
    
    # -- sentiment data
    sent_names = ['date', 'sentiment', 'sentiment_count']
    sentiment = pd.read_excel(path, month, parse_cols='S:U', names=sent_names)
    sentiment = sentiment[sentiment.date.notnull()]
    #sentiment.date = sentiment.date.apply(excel_num2date)
    
    # -- merge all together
    dell_count = radian6[radian6.name == 'Dell']  # retrieve stuff for Dell only from words
    
    count_stats = pd.merge(dell_count, stats, left_on='date', right_on='date')
    data = pd.merge(count_stats, sentiment, left_on='date', right_on='date')
    
    return data, radian6
```

```python
months = ['January', 'February', 'March', 
          'April', 'May', 'June', 
          'July', 'August', 'September', 
          'October', 'November']

# January missing sentiment data
# November is already partially done NO RAW data
pieces_radian6 = []  # list
pieces_data = []  # list

for month in months:
    print 'start', month
    
    data, radian6 = make_stuff_for_month(month)
    
    pieces_data.append(data)
    pieces_radian6.append(radian6)
    print 'finished', month
    
data = pd.concat(pieces_data, ignore_index=True)
data = data.drop_duplicates(cols=['date', 'name', 'sentiment'])

# enforce numerical interpretation of the 
# 'organic reach' and 'viral reach' field 
data['organic reach'] = data['organic reach'].apply(float)
data['viral reach'] = data['viral reach'].apply(float)

radian6 = pd.concat(pieces_radian6, ignore_index=True)
radian6 = radian6.drop_duplicates(cols=['date', 'name'])

# radian6.head(20)
```

```python
more_positive_series = []

def more_positive(group):
    p = group[group.sentiment == 'Positive'].sentiment_count
    n = group[group.sentiment == 'Negative'].sentiment_count
    
    if p > n:
        more_positive_series.append(group[group.sentiment == 'Positive'])
        
    return group
    
data2 = data.groupby('date').apply(more_positive)
```
