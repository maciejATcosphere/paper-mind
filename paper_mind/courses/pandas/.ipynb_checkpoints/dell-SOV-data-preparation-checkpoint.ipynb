```python
social_path = '/home/maciej/projects/temp/SOV_data/social_metrics_and_post_data.xlsx'
SOV_path = '/home/maciej/projects/temp/SOV_data/SOV_and_sentiment.xlsx'

sentiment_multipliers = ({
    'Negative': -2, 
    'Somewhat Negative': -1, 
    'Neutral': 0, 
    'Mixed': 0, 
    'Somewhat Positive': 1,
    'Positive': 2,               
}).items()

multiplier = Series(
    map(lambda x: x[1], sentiment_multipliers), 
    index=map(lambda x: x[0], sentiment_multipliers))
```

```python
from datetime import date, timedelta

def excel_num2date(excel_num):
    if type(excel_num) is datetime:
        return excel_num
    
    return pd.to_datetime(
        date(1900, 1, 1) + timedelta(int(excel_num) - 2))

def collapse_sentiment(g):
    return (g.set_index('sentiment').sentiment_count * multiplier).sum()

# add prop (proportion) column
def add_sov(group):
    # Integer division floors
    count = group.name_count.astype(float)
    
    group['SOV'] = count / count.sum()
    return group
```

```python
def generate_monthly_df(month_index):
    months = ['January', 'February', 'March', 
          'April', 'May', 'June', 
          'July', 'August', 'September', 
          'October', 'November', 'December']
    
    month = months[month_index]
    
    # -- SOV data
    sov_data = pd.read_excel(SOV_path, month, parse_cols='A:C', names=['date', 'name', 'name_count'])
    sov_data = sov_data[sov_data.date.notnull()]  # filter out empty rows 
    sov_data = sov_data.groupby('date').apply(add_sov)
    sov_data.name = sov_data.name.apply(lambda x: x[0].upper() + x[1:])
    sov_data.date = sov_data.date.apply(excel_num2date)
    
    assert np.allclose(sov_data.groupby('date').SOV.sum(), 1)  # sanity check
        
    # -- sentiment data
    sent_names = ['date', 'sentiment', 'sentiment_count']
    sentiment = pd.read_excel(SOV_path, month, parse_cols='S:U', names=sent_names)
    sentiment = sentiment[sentiment.date.notnull()]
    sentiment.date = sentiment.date.apply(excel_num2date)
    sentiment = sentiment[sentiment.date.map(lambda x: x.month - 1) == month_index]
    sentiment = sentiment.groupby('date').agg(collapse_sentiment)[['sentiment_count']].reset_index()
    sentiment.columns = ['date', 'sentiment']
    
    # -- merge all together
    filtered_sov_data = sov_data[sov_data.name == 'Dell']  # retrieve stuff for Dell only from words
    data = pd.merge(filtered_sov_data, sentiment, how='outer', left_on='date', right_on='date')
    
    return data, sov_data

def fetch_sov():
    pieces_sov_data = []
    pieces_data = []
    
    for month_index in range(0, 12):
        data, sov_data = generate_monthly_df(month_index)
        pieces_data.append(data)
        pieces_sov_data.append(sov_data)
    
    data = pd.concat(pieces_data, ignore_index=True)
    sov_data = pd.concat(pieces_sov_data, ignore_index=True)
    
    sov_data = sov_data.drop_duplicates(cols=['date', 'name'])
    data = data.drop_duplicates(cols=['date', 'name'])
    
    return data, sov_data
    
def fetch_insight_stats():
    # fetch insights data
    snames = ["date", "lifetime_likes", "daily_people_talking_about_this", 
              "daily_page_stories", "daily_new_likes", "daily_page_engaged_users", 
              "engagement_rate", "daily_total_reach", "daily_paid_reach", 
              "daily_viral_reach", "daily_total_impressions", 
              "daily_paid_impressions", "daily_viral_impressions", 
              "daily_logged-in_page_views", "daily_total_impressions_of_your_posts", 
              "daily_paid_impressions_of_your_posts", "daily_viral_impressions_of_your_posts", 
              "daily_total_consumers"]
    insight_stats = pd.read_excel(social_path, 'Page Level Data', 
        parse_cols='A:R', names=snames)
    insight_stats = insight_stats[insight_stats.date.notnull()]
    insight_stats.date = insight_stats.date.apply(excel_num2date)
    
    # fetch post stats
    pnames = ["date", "comment", "like", "share"]
    post_stats = pd.read_excel(social_path, 'Post Level Data', 
        parse_cols='A:D', names=pnames)
    post_stats = post_stats[post_stats.date.notnull()]
    post_stats.date = post_stats.date.apply(excel_num2date)
    post_stats = post_stats.fillna(0)
    # in the file there are duplicate entries for the same date
    post_stats = post_stats.groupby('date').sum().reset_index()
    data = pd.merge(insight_stats, post_stats, how='outer', left_on='date', right_on='date')
    
    return data
```

```python
data, sov_data = fetch_sov()
insights_data = fetch_insight_stats()
df = pd.merge(data, insights_data, how='outer', left_on='date', right_on='date')
```
