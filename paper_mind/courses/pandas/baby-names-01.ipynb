```python
import pandas as pd
import numpy as np
```

```python
# read csv
names1880 = pd.read_csv('names/yob1880.txt', names=['name', 'sex', 'births'])
names1880.head()
```

```python
# group with respect to sex column
names1880.groupby('sex').births.sum()
```

```python
# assemble (concatenate) all of the data into single DataFrame
years = range(1880, 2013)

pieces = []  # list
columns = ['name', 'sex', 'births']
for year in years:
    path = 'names/yob{year}.txt'.format(year=year)
    frame = pd.read_csv(path, names=columns)
    
    frame['year'] = year  # add new column `year`
    pieces.append(frame)

names = pd.concat(pieces, ignore_index=True)
names.head()
```

```python
# pivoting the data
total_births = names.pivot_table('births', rows='year', cols='sex', aggfunc=sum)
total_births.tail()
```

```python
# plotting pivot table
total_births.plot(title='Total births by sex and year', figsize=(16, 6));
```

Describe the data:

1. Can you spot any rapid transitions (jumps)?
2. What historical events could be responsible for the shape of the graph?

```python
# add prop (proportion) column
def add_prop(group):
    # Integer division floors
    births = group.births.astype(float)
    
    group['prop'] = births / births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prop)  # TODO: explore in depth apply  

# sanity check of the result
print np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)

names.head()
```

```python
# top 1000 names
def get_top1000(group):
    return group.sort_index(by='births', ascending=False)[:1000]

grouped = names.groupby(['year', 'sex'])
top1000 = grouped.apply(get_top1000)
top1000.head()
```

```python
# analyzing naming trends
total_births = top1000.pivot_table('births', rows='year', cols='name', aggfunc=sum)
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]
subset.plot(subplots=True, figsize=(12, 10), grid=True, title='Number of births per year');
```

```python
# Measuring the increase in naming diversity
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']
table = top1000.pivot_table('prop', rows='year', cols='sex', aggfunc=sum)
table.plot(title='Sum of table1000.prop by year and sex', 
           figsize=(12, 4),
           yticks=np.linspace(0, 1.2, 13), 
           xticks=range(1880, 2020, 10));
```

```python
df = boys[boys.year == 2010] 
prop_cumsum = df.sort_index(by='prop', ascending=False).prop.cumsum()
# print 2010, prop_cumsum.searchsorted(0.5) + 1

df = boys[boys.year == 1900]
in1900 = df.sort_index(by='prop', ascending=False).prop.cumsum()
# print 1900, in1900.searchsorted(0.5) + 1

# in general
def get_quantile_count(group, q=0.5):
    group = group.sort_index(by='prop', ascending=False)
    return group.prop.cumsum().searchsorted(q) + 1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
diversity.plot(title='Number of popular names in top 50%', figsize=(12, 4));
```

```python
# "Last Letter" revolution
# extract last letter from name column
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letter'

table = names.pivot_table('births', rows=last_letters, cols=['sex', 'year'], aggfunc=sum)
subtable = table.reindex(columns=[1910, 1960, 2010], level='year')
subtable.head()
letter_prop = subtable / subtable.sum().astype(float)

import matplotlib.pyplot as plt
fig, axis = plt.subplots(2, 1, figsize=(10, 8))

letter_prop['M'].plot(kind='bar', rot=0, ax=axis[0], title='Male')
letter_prop['F'].plot(kind='bar', rot=0, ax=axis[1], title='Female', legend=False);
```

```python
letter_prop = table / table.sum().astype(float)
dny_ts = letter_prop.ix[['d', 'n', 'y'], 'M'].T
dny_ts.plot()
```

```python
# Boy names that became girl names
all_names = top1000.name.unique()
mask = np.array(['lesl' in x.lower() for x in all_names])
lesley_like = all_names[mask]
filtered = top1000[top1000.name.isin(lesley_like)]
filtered.groupby('name').births.sum()
table = filtered.pivot_table('births', rows='year', cols='sex', aggfunc=sum)
table = table.div(table.sum(1), axis=0)
table.plot(style={'M': 'k-', 'F': 'k--'}, figsize=(12, 4))
```
