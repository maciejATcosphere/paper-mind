```python
import pandas as pd
```

```python
# getting the data
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('./movielens/ml-1m/users.dat', sep='::', header=None, names=unames)

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('./movielens/ml-1m/ratings.dat', sep='::', header=None, names=rnames)

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('./movielens/ml-1m/movies.dat', sep='::', header=None, names=mnames)
```

```python
ratings.head()
```

```python
movies.head()
```

```python
# merging tables together
data = pd.merge(pd.merge(ratings, users), movies)
data.head()
```

```python
# calculating aggregations
mean_ratings = data.pivot_table('rating', rows='title', cols='gender', aggfunc='mean')
mean_ratings.head()
```

```python
# getting group sizes series
ratings_by_title = data.groupby('title').size()
ratings_by_title.head()
```

```python
# filtering some of the not so popular (in terms of amount of ratings) titles
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles[:10]
```

```python
mean_ratings = mean_ratings.ix[active_titles]
mean_ratings.head()
```

```python
# top films among female viewers
top_female_ratings = mean_ratings.sort_index(by='F', ascending=False)
top_female_ratings.head()
```

```python
# measuring rating disagreement
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
sorted_by_diff = mean_ratings.sort_index(by='diff')
sorted_by_diff.head(10)
```

```python
# and those preferred by men
sorted_by_diff[::-1].head()
```

```python
# show those that led to most of disagreement as measured by standard deviation
ratings_std_by_title = data.groupby('title')['rating'].std()
ratings_std_by_title = ratings_std_by_title.ix[active_titles]
ratings_std_by_title.order(ascending=False).head()
```
