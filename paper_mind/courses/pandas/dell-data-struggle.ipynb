```python
from datetime import date, timedelta

def excel_num2date(excel_num):
    return pd.to_datetime(date(1900, 1, 1) + timedelta(excel_num - 2))
```

```python
month = 'February' 
path = '/home/maciej/projects/temp/SOV_data_Final.xlsx'

words = pd.read_excel(path, month, parse_cols='A:C', names=['date', 'name', 'name_count'])
words = words[words.date.notnull()]
words.date = words.date.apply(excel_num2date)

stats = pd.read_excel(path, month, parse_cols='F:P')
stats = stats[stats.Date.notnull()]
stats.Date = stats.Date.apply(pd.to_datetime)

sentiment = pd.read_excel(path, month, parse_cols='S:U')
sentiment = sentiment[sentiment.date.notnull()]
sentiment.date = sentiment.date.apply(excel_num2date)
```

```python
# retrieve stuff for Dell only from words
dell_count = words[words.Word == 'Dell']

count_stats = pd.merge(dell_count, stats, left_on='date', right_on='Date')
count_stats_sentiment = pd.merge(count_stats, sentiment, left_on='date', right_on='date')
```

```python
# add prop (proportion) column
def add_prop(group):
    # Integer division floors
    count = group.name_count.astype(float)
    
    group['prop'] = count / count.sum()
    return group

words = words.groupby('date').apply(add_prop)
grouped = words.groupby('date')

import StringIO

output = StringIO.StringIO()
for name, group in grouped:
    print group
    group.to_csv(output)

# output.getvalue()
```
